```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Top Companies in India 2023.csv')
df
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(df['Revenue(billions US$)'],kde=True)
plt.title('Distribution of Company Revenues')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()
```

```python
top_10 = df.nlargest(10, 'Revenue(billions US$)')
plt.figure(figsize=(12, 8))
sns.barplot(x='Revenue(billions US$)', y='Name', data=top_10, palette='viridis')
plt.title('Top 10 Companies by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Name')
plt.show()
```

```python
industry_revenue = df.groupby('Industry')['Revenue(billions US$)'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x=industry_revenue.values, y=industry_revenue.index, palette='coolwarm')
plt.title('Total Revenue by Industry')
plt.xlabel('Total Revenue')
plt.ylabel('Industry')
plt.show()
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues')
plt.title('Correlation Matrix')
plt.show()
```

```python
grouped_headquarters = df.groupby('Headquarters')['Revenue(billions US$)'].sum().reset_index()
grouped_headquarters = grouped_headquarters.sort_values(by='Revenue(billions US$)', ascending=False)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Headquarters', y='Revenue(billions US$)',data=grouped_headquarters, palette='Set3')
plt.title('Top 10 Headquarter by Revenue')
plt.xlabel('Headquarters')
plt.ylabel('Revenue')
plt.show()
```
